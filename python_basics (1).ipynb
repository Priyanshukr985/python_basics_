```python
# Ques13. Write a Python program to find the sum of all even numbers between 1 and 50.

# AP= 2, 4 , .... 50
a = 2 # first term
d = 2 # common difference ( a1= 1, a2=4 , a2-a1=2)
n = 25  #total 25 terms

sum_even = n/2 * (2*a + (n-1)*d) # sum of first n terms formula
print("Sum of even numbers between 1 and 50 is:", int(sum_even))

```

```python
# Ques14.  Write a program to reverse a string using a while loop.
s = input("Enter a string: ")
reversed_str = ""
index = len(s) - 1

while index >= 0:
    reversed_str += s[index]
    index -= 1

print("Reversed string:", reversed_str)
```

```python
# Ques15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print(f"Factorial of given number is:", factorial)
```
